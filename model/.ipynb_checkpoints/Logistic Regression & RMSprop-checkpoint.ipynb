# Logistic Regression & RMSprop with your own hands

With this notebook we will implement a linear classifier and train it using stochastic gradient descent modifications and numpy.

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

## Two-dimensional classification

To make things more intuitive, let's solve a 2D classification problem with synthetic data.

```python
with open('../data/train.npy', 'rb') as fin:
    X = np.load(fin)
    
with open('../data/target.npy', 'rb') as fin:
    y = np.load(fin)

plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, s=20)
plt.show()
```

## Features

Since that the data above isn't linearly separable, so we should add features (or use non-linear model). Note, that decision line between two classes have the form of circle and we can add quadratic features to make the problem linearly separable. The idea is displayed on the image below:

![](../images/kernel.png)

```python
def expand(X):
    """
    Adds quadratic features. 
    
    For each sample (row in matrix), computes an expanded row:
    [feature0, feature1, feature0^2, feature1^2, feature1*feature2, 1]
    
    :param X: matrix of features, shape [n_samples,2]
    :returns: expanded features of shape [n_samples,6]
    """
    
    X_expanded = np.zeros((X.shape[0], 6))
    
    return np.stack((X[:, 0], X[:, 1], X[:, 0] ** 2, X[:, 1] ** 2, X[:, 0] * X[:, 1], np.ones_like(X[:, 0])), axis=1)
```

```python
X_expanded = expand(X)
X_expanded.shape
```

And before we start, let's divide data into train and test. Data is already shuffled, so we can do it just like that:

```python
count_train_examples = 600 #can experiment with this
X_expanded_train = X_expanded[:count_train_examples, :]
X_expanded_test = X_expanded[count_train_examples:, :]
y_train = y[:count_train_examples]
y_test = y[count_train_examples:]
```

```python
print(X_expanded_train.shape, X_expanded_test.shape)
```

## Logistic regression

To classify objects we will obtain probability of object belongs to class '1'. To predict probability we will use output of linear model and logistic function:

$$ a(x; w) = \langle w, x \rangle $$
$$ P( y=1 \; \big| \; x, \, w) = \dfrac{1}{1 + \exp(- \langle w, x \rangle)} = \sigma(\langle w, x \rangle)$$


```python
def sigma(X, w):
    """
    Returns predicted probabilities y==1 of given x, P(y=1|x)
    
    :param X: feature matrix X of shape [n_samples,6] (expanded)
    :param w: weight vector w of shape [6] for each of the expanded features
    :returns: an array of predicted probabilities in [0,1] interval.
    """

    return 1 / (1 + np.exp(-np.dot(X, w)))
```

In logistic regression the optimal parameters $w$ are found by cross-entropy minimization:

$$ L(w) =  - {1 \over \ell} \sum_{i=1}^\ell \left[ {y_i \cdot log P(y_i \, | \, x_i,w) + (1-y_i) \cdot log (1-P(y_i\, | \, x_i,w))}\right] $$



```python
def compute_loss(X, y, w):
    """
    Given feature matrix X [n_samples,6], target vector [n_samples] of 1/0,
    and weight vector w [6], computes scalar loss function using formula above.
    """
    
    y_pred = sigma(X, w)
    f1 = np.dot(y, np.log(y_pred))
    f2 = np.dot(1 - y, np.log(1 - y_pred))
    fac = -1 / y.shape[0]
    res = fac * (f1 + f2)
    
    return res
```

Since we train our model with gradient descent, we should compute gradients.

To be specific, we need a derivative of loss function over each weight [6 of them].

$$ \nabla_w L(w) = - {1 \over \ell} \sum_{i=1}^\ell y_i \dfrac{\nabla_w \sigma(\langle w, x_i \rangle)}{\sigma(\langle w, x_i \rangle)} + (1-y_i) \dfrac{- \nabla_w \sigma(\langle w, x_i \rangle)}{1 - \sigma(\langle w, x_i \rangle)} = ... = - {1 \over \ell} \sum_{i=1}^\ell \dfrac{y_i \nabla_w \sigma(\langle w, x_i \rangle) - \sigma(\langle w, x_i \rangle) \nabla_w \sigma(\langle w, x_i \rangle)}{\sigma(\langle w, x_i \rangle) (1 - \sigma(\langle w, x_i \rangle))} $$

If we turn to matrix operations:

$$ \nabla_w L(w) = - {1 \over \ell} \dfrac{y^{\text{T}} \nabla_w \sigma(\langle X, w \rangle) - \sigma(\langle X, w \rangle)^{\text{T}} \nabla_w \sigma(\langle X, w \rangle)}{\sigma(\langle X, w \rangle)^{\text{T}} (1 - \sigma(\langle X, w \rangle))} $$


```python
def dPdW(X, w):
    res = np.dot(np.dot((1 - sigma(X, w)), sigma(X, w)), X)
    return res

def compute_grad(X, y, w):
    """
    Given feature matrix X [n_samples,6], target vector [n_samples] of 1/0,
    and weight vector w [6], computes vector [6] of derivatives of L over each weights.
    """
    
    l = len(y)
    y_pred = sigma(X, w)
    dpdw = dPdW(X, w)
    res = (np.dot(y, dpdw) - np.dot(y_pred, dpdw)) / np.dot(y_pred, 1 - y_pred) / (-l)
    
    return res
```

Here's an auxiliary function that visualizes the predictions:

```python
from IPython import display

h = 0.01
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

def visualize(X, y, w, history):
    """draws classifier prediction with matplotlib"""
    Z = sigma(expand(np.c_[xx.ravel(), yy.ravel()]), w)
    Z = Z.reshape(xx.shape)
    plt.figure(figsize=(15, 6))
    plt.subplot(1, 2, 1)
    plt.contourf(xx, yy, Z, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.colorbar()
    
    plt.subplot(1, 2, 2)
    plt.plot(history)
    plt.grid()
    ymin, ymax = plt.ylim()
    plt.ylim(0, ymax)
    display.clear_output(wait=True)
    plt.show()
```

## Training
In this section we'll use the functions we wrote to train our classifier using stochastic gradient descent.

You can try change hyperparameters like batch size, learning rate and so on to find the best one.

## Mini-batch SGD

Stochastic gradient descent just takes a random example on each iteration, calculates a gradient of the loss on it and makes a step:
$$ w_t = w_{t-1} - \eta \dfrac{1}{m} \sum_{j=1}^m \nabla_w L(w_t, x_{i_j}, y_{i_j}) $$



```python
np.random.seed(42)
w = np.array([0, 0, 0, 0, 0, 1])

eta = 0.1 # learning rate

n_iter = 200
batch_size = 20
loss = np.zeros(n_iter)
plt.figure(figsize=(12, 5))

for i in range(n_iter):
    ind = np.random.choice(X_expanded_train.shape[0], batch_size)
    loss[i] = compute_loss(X_expanded_train, y_train, w)
    
    if i % 10 == 0:
        visualize(X_expanded_train[ind, :], y_train[ind], w, loss)
        
    w = w - eta * compute_grad(X_expanded_train[ind, :], y_train[ind], w)

visualize(X, y, w, loss)
plt.clf()
```

The result looks good: our model has caught the concept of separation. 

Notice, that the loss curve hasn't reached the asymptote yet! So, we can do better. 

## SGD with momentum

Momentum is a method that helps accelerate SGD in the relevant direction and dampens oscillations as can be seen in image below. It does this by adding a fraction $\alpha$ of the update vector of the past time step to the current update vector.
<br>
<br>

$$ \nu_t = \alpha \nu_{t-1} + \eta\dfrac{1}{m} \sum_{j=1}^m \nabla_w L(w_t, x_{i_j}, y_{i_j}) $$
$$ w_t = w_{t-1} - \nu_t$$

<br>


![](../images/sgd.png)


```python
np.random.seed(42)
w = np.array([0, 0, 0, 0, 0, 1])

eta = 0.1 # learning rate
alpha = 0.9 # momentum
nu = np.zeros_like(w)

n_iter = 200
batch_size = 20
loss = np.zeros(n_iter)
plt.figure(figsize=(12, 5))

for i in range(n_iter):
    ind = np.random.choice(X_expanded_train.shape[0], batch_size)
    loss[i] = compute_loss(X_expanded_train, y_train, w)
    
    if i % 10 == 0:
        visualize(X_expanded_train[ind, :], y_train[ind], w, loss)
        
    nu = alpha * nu + eta * compute_grad(X_expanded_train[ind, :], y_train[ind], w)
    w = w - nu

visualize(X, y, w, loss)
plt.clf()
```

It seems much better! The model area of doubt became less and we could reduce loss value.

Can notice, that algorithm converges much faster.
Loss curve has reached the asymptote on the nearly 150-th iteration.

## RMSprop

Implement RMSPROP algorithm, which use squared gradients to adjust learning rate:

$$ G_j^t = \alpha G_j^{t-1} + (1 - \alpha) g_{tj}^2 $$
$$ w_j^t = w_j^{t-1} - \dfrac{\eta}{\sqrt{G_j^t + \varepsilon}} g_{tj} $$

```python
np.random.seed(42)
w = np.array([0, 0, 0, 0, 0, 1.])

eta = 0.05 # learning rate
alpha = 0.9 # moving average of gradient norm squared
g2 = np.zeros_like(w)
eps = 1e-8

n_iter = 200
batch_size = 20
loss = np.zeros(n_iter)
plt.figure(figsize=(12,5))
for i in range(n_iter):
    ind = np.random.choice(X_expanded_train.shape[0], batch_size)
    loss[i] = compute_loss(X_expanded_train, y_train, w)

    if i % 10 == 0:
        visualize(X_expanded_train[ind, :], y_train[ind], w, loss)
    
    grad = compute_grad(X_expanded_train[ind, :], y_train[ind], w)
    g2 = alpha * g2 + (1 - alpha) * grad ** 2
    w = w - eta / np.sqrt(g2 + eps) * grad

visualize(X, y, w, loss)
plt.clf()
```

Almost the same result.

Let's check the accuracy score:

```python
def accuracy_score(X_test, y_test, w):
    n = len(y_test)
    
    y_probs = sigma(X_test, w)
    prob_threshold = 0.5 #you can experiment with this
    y_pred = [1 if y_probs[i] > prob_threshold else 0 for i in range(n)]
    
    return sum([1 if y_test[i] == y_pred[i] else 0 for i in range(n)]) / n
```

```python
accuracy_score(X_expanded_test, y_test, w)
```
